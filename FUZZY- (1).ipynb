```python
import numpy as np
import matplotlib.pyplot as plt

#    L   
def L1(x, m, alpha, p):
    return np.maximum(0, 1 - np.power(np.abs((m-x[0])/alpha), p))


def L2(x, m, alpha, p):
    return np.exp(-np.power(np.abs((m-x[0])/alpha), p))

def L3(x, m, alpha, p):
    return 1 / (1 + np.power(np.abs((m-x[0])/alpha), p))

def L4(x, m, alpha, p):
    return 1 / (1 + p*np.abs((m-x[0])/alpha))

#    R   
def R1(x, m, beta, p):
    return np.maximum(0, 1 - np.power(np.abs((x[0]-m)/beta), p))

def R2(x, m, beta, p):
    return np.exp(-np.power(np.abs((x[0]-m)/beta), p))

def R3(x, m, beta, p):
    return 1 / (1 + np.power(np.abs((x[0]-m)/beta), p))

def R4(x, m, beta, p):
    return 1 / (1 + p*np.abs((x[0]-m)/beta))

#    
m = float(input("Enter a fuzzy number m: "))
alpha = float(input("Enter the left limit alpha: "))
beta = float(input("Enter the right limit beta: "))
p = float(input("Enter the parameter p: "))

#   L  R   
x_values = np.linspace(0, 10, 100)
for i, L in enumerate([L1, L2, L3, L4]):
    for j, R in enumerate([R1, R2, R3, R4]):
        y_values = np.maximum(np.minimum(L((x_values,), m, alpha, p), 1), np.clip(R((x_values,), m, beta, p), 0, 1))
        plt.subplot(4, 4, i*4+j+1)
        plt.plot(x_values, y_values)
        plt.title('L{}R{}'.format(i+1, j+1))
        plt.xlabel('x')
        plt.ylabel('Membership degree')

plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```
