```python
import numpy as np
import matplotlib.pyplot as plt

def fuzzy_number(m, alpha, beta, p, l, r):
    L = m - alpha * l
    R = m + beta * r
    return L, R

def plot_fuzzy_number(L, R, p, l, r):
    x = np.linspace(L - p*l, R + p*r, 1000)
    y = np.zeros_like(x)

    for i, xi in enumerate(x):
        if xi < L:
            y[i] = max(0, 1 - (L - xi) / (p * l))
        elif xi > R:
            y[i] = max(0, 1 - (xi - R) / (p * r))
        else:
            y[i] = 1

    plt.plot(x, y)
    plt.xlabel('X')
    plt.ylabel('Membership Degree')
    plt.title('Fuzzy Number')
    plt.show()

m = float(input("Enter m: "))
alpha = float(input("Enter alpha: "))
beta = float(input("Enter beta: "))
p = float(input("Enter p: "))
l = float(input("Enter L: "))
r = float(input("Enter R: "))

L, R = fuzzy_number(m, alpha, beta, p, l, r)
plot_fuzzy_number(L, R, p, l, r)
```
